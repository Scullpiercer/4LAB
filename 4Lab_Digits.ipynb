```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digit['data'], digit['target'], random_state=0)
```

```python
print("Форма массива X_train:{}".format(X_train.shape))
print("Форма массива y_train:{}".format(y_train.shape))
```

```python
print("Форма массива X_test:{}".format(X_test.shape))
print("Форма массива y_test:{}".format(y_test.shape))
```

```python
digit_df = pd.DataFrame(digit.data, columns=digit.feature_names)
digit_df['target'] = digit.target
digit_df.head()
```

```python
digits_dataframe = pd.DataFrame(X_train[:49, :7], columns=digit.feature_names[:7])
grr = pd.plotting.scatter_matrix(digits_dataframe, c=y_train[:49], figsize=(15, 15), marker='o', hist_kwds={'bins': 20}, alpha=.8)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train, y_train)
```

```python
X_new = np.array([[0, 2, 9, 15, 14, 9, 3, 0, 0, 4, 13, 8, 9, 16, 8, 0, 0, 0,
0, 6, 14, 15, 3, 0, 0, 0, 0, 11, 14, 2, 0, 0, 0, 0, 0, 2,
15, 11, 0, 0, 0, 0, 0, 0, 2, 15, 4, 0, 0, 1, 5, 6, 13, 16,
6, 0, 0, 2, 12, 12, 13, 11, 0, 0
]])
print("Форма массива X_new: {}".format(X_new.shape))
```

```python
prediction = knn.predict(X_new)
print("Прогноз: {}".format(prediction))
print("Спрогнозированнная метка: {}".format(digit['target_names'][prediction]))
```

```python
y_pred = knn.predict(X_test)
print("Прогнозы для тестового набора:\n {}".format(y_pred))
```

```python
print("Правильность на тестовом наборе: {:.2f}".format(np.mean(y_pred == y_test)))
```

```python
print("Правильность на тестовом наборе: {:.2f}".format(knn.score(X_test, y_test)))
```
